```python
import cv2
import numpy as np
import random
windowwidth = 400
windowlength = 600
windowwidth_mid = windowwidth/2 #200
windowlength_mid = windowlength/2 #300
img_2= np.zeros((windowwidth,windowlength,3), dtype='uint8') #400 y長度,x長度,3維度
#座標(0,0):200,300: (int(img.shape[1]/2) ,int(img.shape[0]/2) )

'''建立座標集物件'''
class colist:#輸入座標
    def __init__(self):
        self.co_list=[]
        self.cox=[]
        self.coy=[]
    def colist_input(self):
        while True:#重複值確認
            i=1
            while True:
                each_co=[]
                try:
                    print("X",i,":",sep="",end="")
                    each_co_x = int(input())
                    print("Y",i,":",sep="",end="")
                    each_co_y =int(input())
                    each_co.append(each_co_x)
                    each_co.append(each_co_y)
                    self.cox.append((each_co_x)*20)#數字太小蕙沒辦法判斷&畫圖
                    self.coy.append((each_co_y)*20)
                    self.co_list.append(each_co)
                    i+=1
                    if i==5:
                        break
                except:
                    print("Input Error")
            if self.colist_check():
                break
            else:
                self.co_list=[]
    def colist_check(self):
        a=""
        for i in self.co_list:
            a+=(str(i[0])+str(i[1])+",")
        #a.split(",")
        if len(set(a.split(",")))!=5:
            print("有重複值")
            return False
        else:
            print("ok")
            for i in self.co_list:
                print("(",i[0],",",i[1],")",sep="",end=" ")
            return True
'''四邊形判斷FUNCTION'''
def quadrilateral():
    ANS=""
    co_list=colistA.co_list
    sorted(co_list)
    LL=[] #Line_List

    line_1 = (abs(co_list[0][0]-co_list[1][0])**2 + abs(co_list[0][1]-co_list[1][1])**2)**(1/2)
    line_2 = (abs(co_list[1][0]-co_list[2][0])**2 + abs(co_list[1][1]-co_list[2][1])**2)**(1/2)
    line_3 = (abs(co_list[2][0]-co_list[3][0])**2 + abs(co_list[2][1]-co_list[3][1])**2)**(1/2)
    line_4 = (abs(co_list[3][0]-co_list[0][0])**2 + abs(co_list[3][1]-co_list[0][1])**2)**(1/2)
    line_5 = (abs(co_list[1][0]-co_list[3][0])**2 + abs(co_list[1][1]-co_list[3][1])**2)**(1/2)
    line_6 = (abs(co_list[0][0]-co_list[2][0])**2 + abs(co_list[0][1]-co_list[2][1])**2)**(1/2)

    LL.append(line_1)
    LL.append(line_2)
    LL.append(line_3)
    LL.append(line_4)
    LL.append(line_5)
    LL.append(line_6)
    LL.sort()
    
    if LL[0]==LL[1] and LL[1]==LL[2] and LL[2]==LL[3] and LL[0]==LL[3]:        
        if round(LL[0] * (2**(1/2)),6) == round(LL[5],6) : #1:1:2(1/2)
            ANS="Square"
    elif LL[1]==LL[2] and LL[2]==LL[3] and LL[3]==LL[4]:
        ANS="Diamond"
    elif (LL[0]==LL[1] and LL[2]==LL[3]) and (LL[4]==LL[5]) and (round((LL[0]**2 + LL[2]**2),6) == round(LL[5]**2),6): #長方形
        ANS="Rectangle"
    elif ((((LL[1]**2)-((LL[2]*0.5)**2))**(1/2))+(((LL[3]**2)-((LL[2]*0.5)**2))**(1/2)) ==LL[5] )or((((LL[1]**2)-((LL[0]*0.5)**2))**(1/2))+(((LL[3]**2)-((LL[0]*0.5)**2))**(1/2)) ==LL[5] ):
        ANS = "Kite"
    elif (LL[0]==LL[1] or LL[1]==LL[2] )and LL[3]==LL[4] :
        ANS="Parallelogram OR Trapezoid"
    else:
        ANS="UNREGULAR"
    return ANS

'''繪製圖形'''
while True: #判斷輸入座標是否能形成四邊形
    colistA=colist()
    colistA.colist_input() 
    print()
    pts = np.array([[windowlength_mid+colistA.cox[0],windowwidth_mid-colistA.coy[0]],
                    [windowlength_mid+colistA.cox[1],windowwidth_mid-colistA.coy[1]],
                    [windowlength_mid+colistA.cox[2],windowwidth_mid-colistA.coy[2]],
                    [windowlength_mid+colistA.cox[3],windowwidth_mid-colistA.coy[3]]],dtype='uint64')   # 產生座標陣列
    count=0
    print("圖形繪製中",end="")
    while True:#處理四個點形成四邊形的連線方式-凸四邊形y
        img_2 = np.zeros((400,600,3), dtype='uint8')
        new_pts = np.zeros((4,2), dtype='uint64')
        coorder = random.sample(range(0,4),4)
        for i,j in zip(coorder,range(4)):
            new_pts[j]+=pts[i] #產生新順序了
        cv2.polylines(img_2,[new_pts],True,(0,0,255),2) #在img_2上畫多邊形，不被座標軸影響：true閉合
        canny = cv2.Canny(img_2,1,10)#邊緣
        contours,hierarchy = cv2.findContours(canny,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)#圖片,(外、內、都要),壓縮 [回傳判斷為輪廓的所有點&階層(用不到)]
        for cnt in contours:
            peri = cv2.arcLength(cnt,True)#輪廓邊常 ,T/F輪廓是否閉合
            vertices = cv2.approxPolyDP(cnt,peri*0.02,True) #近似形狀
        print(".",end="")
        #print("角數量:",len(vertices),end="")
        #print("面積",area)
        if len(vertices)==4: #判斷是否是四邊形
            break
        count+=1
        if count>20:
            print("輸入之座標無法形成四邊形，重新輸入")
            break
    if len(vertices)==4:
        break
"""畫四個點、標記座標、XY軸"""
for i,j in zip(range(4),new_pts):#中心座標要用int表示
    cv2.circle(img_2,(int(new_pts[i][0]),int(new_pts[i][1])),2,(255,255,255),cv2.FILLED)      # 圓：要畫的圖片,中心點,半徑,顏色,線條粗度
    name = int((int(j[0])-300)/20),int((200-int(j[1]))/20)
    name = str(name)
    cv2.putText(img_2,name,(int(new_pts[i][0]),int(new_pts[i][1])),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),1)
cv2.line(img_2,(int(img_2.shape[1]/2),0),(int(img_2.shape[1]/2),img_2.shape[0]),(255,255,255),1) #X軸
cv2.line(img_2,(0,int(img_2.shape[0]/2)),(img_2.shape[1],int(img_2.shape[0]/2)),(255,255,255),1) #Y軸

"""判斷形狀!"""
print("\nINTRO:\nWhat kind of quadrilateral can a Given list of four coordinates constitude：")
ANS=quadrilateral()
cv2.putText(img_2,ANS,(2,windowwidth-10),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),1)
print(ANS)


cv2.imwrite('Draw_A_Picture.jpg', img_2)        # 存成 jpg
cv2.imshow('DrawWindow', img_2)
cv2.waitKey(0)==ord("q")                             # 按下任意鍵停止
cv2.destroyAllWindows()
```
